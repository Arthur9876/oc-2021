## Créer un nouveau Git local

    % git init myproject
    Dépôt Git vide initialisé dans /Users/raphael/oc/myproject/.git/
    (base) raphael@MacBook-Air-de-Raphael oc % 

Aller dans le nouveau projet

    % cd myproject 

Afficher le fichier caché `.git` qui contient la gestion du projet.

    % ls -a
    .	..	.git

Afficher le contenu de ce dossier caché 

    % ls .git
    HEAD		description	info		refs
    config		hooks		objects

La commande `tree` permet d'afficher la structure:

    % tree .git

## Configuration

Affichons le fichier de configuration inital

    % cat .git/config 
    [core]
        repositoryformatversion = 0
        filemode = true
        bare = false
        logallrefupdates = true
        ignorecase = true
        precomposeunicode = true

Ajoutons le nom d'utilisateur

    % git config user.name 'Raphael Holzer'

Pour vérifier

    % git config user.name
    Raphael Holzer

Ajoutons aussi l'email 

    % git config user.email 'raphael.holzer@eduvaud.ch'

Affichons le contenu du fichier `config`. Nous constatons que deux lignes ont été ajouté.

    cat .git/config     
    [core]
        repositoryformatversion = 0
        filemode = true
        bare = false
        logallrefupdates = true
        ignorecase = true
        precomposeunicode = true
    [user]
        name = Raphael Holzer
        email = raphael.holzer@eduvaud.ch


## Créer un fichier

Nous ajoutons un nouveau ficher, une page web écrit en HTML.

    nano index.html

Entrez le texte

    <!DOCTYPE html>
    <html>
    <h1> Ma page web </h1>
    La toute premier ligne.
    </html>

Sauvegardons et vérifions le status.

    % git status
    Sur la branche master

    Aucun commit

    Fichiers non suivis:
    (utilisez "git add <fichier>..." pour inclure dans ce qui sera validé)

        index.html

    aucune modification ajoutée à la validation mais des fichiers non suivis sont présents (utilisez "git add" pour les suivre)




## Ajouter le fichier

Nous devons maintenant ajouter le fichier à git.

    % git add index.html 

Avec `git status` nous pouvons de nouveau vérifier le status.

    % git status
    Sur la branche master

    Aucun commit

    Modifications qui seront validées :
    (utilisez "git rm --cached <fichier>..." pour désindexer)

        nouveau fichier : index.html


Nous déposons le fichier avec un message explicatif.

    % git commit -m 'inital commit'
    [master (commit racine) 18f8adc] inital commit
    1 file changed, 2 insertions(+)
    create mode 100644 index.html

## Afficher la page

Vous pouvez ouvrir la page avec un navigateur avec la commande

    open index.html

L'extension .html indique au système de lancer un navigateur web et d'afficher la page.

## Modifier le fichier

 Faison la correction 
 
    % git commit -a -m 'correct HTML'
    [master 043af4c] correct HTML
    1 file changed, 5 insertions(+), 2 deletions(-)
